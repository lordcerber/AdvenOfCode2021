```python
test_input = """7,4,9,5,11,17,23,2,0,14,21,24,10,16,13,6,15,25,12,22,18,20,8,19,3,26,1

22 13 17 11  0
 8  2 23  4 24
21  9 14 16  7
 6 10  3 18  5
 1 12 20 15 19

 3 15  0  2 22
 9 18 13 17  5
19  8  7 25 23
20 11 10 24  4
14 21 16 12  6

14 21 17 24  4
10 16 15  9 19
18  8 23 26 20
22 11 13  6  5
 2  0 12  3  7""".splitlines();
```

```python
def read_input (input):
    sequence = list(map(int, input[0].split(",")))
    start=2
    bingos = []
    while start+5<=len(input):
        bingos.append([list(map(int, (i.split()))) for i in input[start:start+5]])
        start+=6
    return sequence, bingos
```

```python
import numpy as np

def check_bingo (b):
    bingo=np.array(b);
    if (min(np.concatenate((np.sum(bingo, axis=0),np.sum(bingo, axis=1)), axis=0))==-5):
        return 1
    return 0
```

```python
import numpy as np
def play_bingo (input):
    seq, b = read_input(input)
    bingos = np.array(b)
    for a in seq:
        print(a, bingos)
        bingos = np.where(bingos==a, -1, bingos)
        print(a,bingos)
        for bingo in bingos:
            if check_bingo (bingo):
                print("BINGO")
                print (bingo)
                sum = np.sum(np.where(bingo==-1,0,bingo))
                print (sum)
                return sum*a

        
```

```python
import numpy as np
def play_bingo_last (input):
    seq, b = read_input(input)
    bingos = np.array(b)
    for a in seq:
        print(a, bingos)
        bingos = np.where(bingos==a, -1, bingos)
        print(a,bingos)
        if len(bingos)>1:
            bingos =np.array([b for b in bingos if not check_bingo(b)])
        elif check_bingo (bingos[0]):
                bingo=bingos[0]
                print("LAST BINGO")
                print (bingo)
                sum = np.sum(np.where(bingo==-1,0,bingo))
                print (sum)
                return sum*a
```
